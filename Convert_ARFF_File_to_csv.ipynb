```python
from scipy.io import arff
import pandas as pd
```

```python
# df = pd.read_csv("Data/winequality-red.csv")
data, meta = arff.loadarff("Data/Training Dataset.arff")
```

```python
df = pd.DataFrame(data)
```

```python
for col in df.select_dtypes(include=["object"]):
    # If values are bytes, decode them
    if isinstance(df[col].iloc[0], (bytes, bytearray)):
        df[col] = df[col].str.decode("utf-8", errors="ignore")
```

```python
df.to_csv("Data/phishing_data.csv", index=False)
```
